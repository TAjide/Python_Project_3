```python
#Creating a subset of the columns that i want to keep

df2 = df[['ARREST_KEY', 'ARREST_DATE', 'PD_DESC', 'OFNS_DESC'\
          ,'LAW_CAT_CD', 'ARREST_BORO', 'ARREST_PRECINCT', 'JURISDICTION_CODE'\
          ,'AGE_GROUP', 'PERP_SEX', 'PERP_RACE']]

df2.head()
```

```python
#importing Libraries for analysis

import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#checking to see where the missing data are located

missing = np.where(df2['LAW_CAT_CD'].isnull() == True)
missing
```

```python
#Count by Law category: Misdemeanor, Felony, Violation

df2['LAW_CAT_CD'].value_counts()
```

```python
#Getting an idea of what values are in the missing locations

df2.loc[61911:, 'LAW_CAT_CD'].head(15)
```

```python
####Removing missing + unknown data

df1 = df2[(df2['LAW_CAT_CD'] != '9') & (df2['LAW_CAT_CD'] != 'I')\
          & (df2['LAW_CAT_CD'] != 'NaN') & (df2['LAW_CAT_CD'] != '(null)')\
          & (df2['PERP_RACE'] != 'UNKNOWN')]

df1.head()
```

```python
#Visualization on overall data

sns.pairplot(df1, hue='PERP_SEX')
```

```python
#Count by Law category: Misdemeanor, Felony, Violation
#Confirming that unwanted rows were dropped

df1['LAW_CAT_CD'].value_counts()
```

```python
df1.shape
```

```python
##a quick look at the information about dataframe

df1.info()
```

```python
#All at once--- summarized descriptive satistics

df1.describe()
```

```python
####Count by age group

df1['AGE_GROUP'].value_counts()
```

```python
#visualization of data by age group----- ages 25-44 have the highest number of arrests

df1['AGE_GROUP'].value_counts() \
.head(10) \
.plot(kind='bar', xlabel = 'AGE_GROUP', rot=45,
      ylabel = 'Overall count', color = 'coral')
plt.title('Age Group Comparison on Overall count' \
          , loc='center', color='olive', style='italic')
plt.show()
```

```python
####Count by gender---- Males have the higher arrest counts

df1['PERP_SEX'].value_counts()
```

```python
#visualization of data by gender and arrest counts

df1['PERP_SEX'].value_counts() \
.head(10) \
.plot(kind='bar', xlabel = 'Gender', rot=45,
      ylabel = 'Overall count', color = 'coral')
plt.title('Gender Comparison on Overall count' \
          , loc='center', color='olive', style='italic')
plt.show()
```

```python
# distribution, by gender + age group

grouping_by_gender = df1.groupby(['PERP_SEX', 'AGE_GROUP']).size().reset_index(name='count')

grouping_by_gender
```

```python
###Visualization of data by gender + age group

plt.figure(figsize=(7,5))
sns.barplot(x='PERP_SEX',y='count',palette='Wistia', hue='AGE_GROUP' \
            , data=grouping_by_gender,color='coral_r')
plt.title('$\mathbf{Gender Comparison on age group .vs. Gender}$' \
          , loc='center', color='coral')
plt.xlabel('Gender')
plt.ylabel('Total Count')
plt.legend(loc='upper right')

plt.show()
```

```python
#Gender distribution by level of offense--- Males with Misdemeanor has highest count

grouping_by_offense = df1.groupby(['PERP_SEX', 'LAW_CAT_CD']).size().reset_index(name='count')

grouping_by_offense
```

```python
###Visualization of data by gender + level of offense

plt.figure(figsize=(5,7))
sns.barplot(x='PERP_SEX',y='count',palette='Set1_r', hue='LAW_CAT_CD' \
            , data=grouping_by_offense,color='olive_r')
plt.title('$\mathbf{Gender Comparison on Level of offense}$' \
          , loc='center', color='coral')
plt.xlabel('Gender')
plt.ylabel('Total Count')
plt.legend(loc='upper right')

plt.show()
```

```python
# distribution by gender + race

grouping_by_race = df1.groupby(['PERP_SEX', 'PERP_RACE']).size().reset_index(name='count')

grouping_by_race
```

```python
###Visualization of data by gender + race

plt.figure(figsize=(9,7))
sns.barplot(x='PERP_SEX',y='count', palette='autumn', hue='PERP_RACE' \
            , data=grouping_by_race,color='pink_r')
plt.title('$\mathbf{Gender Comparison by race}$' \
          , loc='center', color='coral')
plt.xlabel('Gender')
plt.ylabel('Total Count')
plt.legend(loc='upper left')

plt.show()
```

```python
# distribution by level of offense + race

grouping_by_race_offense = df1.groupby(['LAW_CAT_CD', 'PERP_RACE']).size().reset_index(name='count')

grouping_by_race_offense
```

```python
###Visualization of data by level of offense (Felony, Misdemeanor, Violation) + race

plt.figure(figsize=(8,7))
sns.barplot(x='LAW_CAT_CD',y='count',palette='PuBu', hue='PERP_RACE' \
            , data=grouping_by_race_offense,color='yellow_r')
plt.title('$\mathbf{Race Comparison by level of offense}$' \
          , loc='center', color='coral')
plt.xlabel('Offense Category')
plt.ylabel('Total Count')
plt.legend(loc='upper right')

plt.show()
```

```python
# distribution, by race + age group

grouping_by_age_race = df1.groupby(['PERP_RACE', 'AGE_GROUP']).size().reset_index(name='count')

grouping_by_age_race
```

```python
####Violin plot of distribution by age group

figsize = (12, 1.2 * len(grouping_by_age_race['AGE_GROUP'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(grouping_by_age_race, x='count', y='AGE_GROUP', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
###Visualization of data by age group + race---ages 25-44 has the highest arrest count in race groups

plt.figure(figsize=(17,5))
sns.barplot(x='PERP_RACE',y='count',palette='Dark2',
            hue='AGE_GROUP' \
            , data=grouping_by_age_race,color='yellow_r')
plt.title('$\mathbf{Race Comparison by age Group}$' \
          , loc='center', color='coral')
plt.xlabel('Race')
plt.ylabel('Total Count')
plt.legend(loc='upper right')

plt.show()
```

```python
###grouping by borough_race_gender

grouping_by_boro_race_gend = df1.groupby(['ARREST_BORO', 'PERP_RACE', 'PERP_SEX']).size().reset_index(name='count')

grouping_by_boro_race_gend
```

```python
####Violin plot of distribution by race

figsize = (12, 1.2 * len(grouping_by_boro_race_gend['PERP_RACE'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(grouping_by_boro_race_gend, x='count', y='PERP_RACE', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)

plt.show()
```

```python
#Violin plot of distribution by Borough---- K(Broklyn) has the highest arrest count

figsize = (12, 1.2 * len(grouping_by_boro_race_gend['ARREST_BORO'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(grouping_by_boro_race_gend, x='count', y='ARREST_BORO', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)

plt.show()
```

```python
# distribution, by Borough + gender

grouping_by_boro_gender = df1.groupby(['PERP_SEX', 'ARREST_BORO']).size().reset_index(name='count')

grouping_by_boro_gender
```

```python
###Visualization of data by Borough and Gender

plt.figure(figsize=(7,5))
sns.barplot(x='PERP_SEX',y='count', palette='afmhot', hue='ARREST_BORO' \
            , data=grouping_by_boro_gender,color='peru_r')
plt.title('$\mathbf{Gender Comparison by Borough}$' \
          , loc='center', color='coral')
plt.xlabel('Gender')
plt.ylabel('Total Count')
plt.legend(loc='upper right')

plt.show()
```

```python
# distribution, by Borough + Race

grouping_by_race_boro = df1.groupby(['PERP_RACE', 'ARREST_BORO']).size().reset_index(name='count')

grouping_by_race_boro
```

```python
###Visualization of data by Borough and Race

plt.figure(figsize=(17,5))
sns.barplot(x='PERP_RACE',y='count', palette='Paired', hue='ARREST_BORO' \
            , data=grouping_by_race_boro,color='coral_r')
plt.title('$\mathbf{Race Comparison by Borough}$' \
          , loc='center', color='coral')
plt.xlabel('Race')
plt.ylabel('Total Count')
plt.legend(loc='upper right')

plt.show()
```

Analysis show that ages 25-44 has the highest arrest count among both males and female in overall count and across race groups, with males having the highest arrest count overall. Males with misdemeanor have the highest count in arrest types and next were males with felony arrests. This same trend also appears in the female data count, with higher misdemeanor count and felony arrests running second.
Black males have the highest overall count with majority of arrests from Brooklyn Borough. Confirmation or rejection of bias was unable to be determined, due to no current census information on Borough. The most current data available on census count for Brooklyn was in 2020, which may potentially skew the result of anaysis on possible bias or lack thereof.
